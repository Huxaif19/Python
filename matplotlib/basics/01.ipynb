```python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [4,5,6]
plt.scatter(x,y)
```

```python
plt.plot(x,y, "b--")    # b for the color and -- for the line type can be p, ^ etc.
plt.scatter(x,y, color = "black")
plt.show()
```

plotting a function (x^3)

```python
import numpy as np

# x = np.array([1,2,3,4])
x = np.arange(0,5,0.1)
y = x ** 3
plt.plot(x,y , "rp")
```

```python
a = [1,2,3]   
plt.plot(a, "b--")    # here a = y , x i treated as range(n)
```

```python
x = np.arange(0,10,0.1)
y = x ** 3
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Matplotlib demo")
plt.plot(x,y , color = "black",marker = "", linewidth = 4, label = "X^3")
plt.plot(x, x ** 2, color = "red", marker = "", linewidth = 4, label = "X^2")
plt.legend()
plt.axis([0, 10, 0, 200])
plt.grid()
plt.text(8, 64, "square", fontsize = 8)
```
