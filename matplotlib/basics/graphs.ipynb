```python
import matplotlib.pyplot as plt
import numpy as np


year       = [2012, 2013,2014,2015,2016, 2017]
salary     = [12,13,14,15,16,17]
population = [100,120,180, 250,300,370]
color = np.arange(len(year))
```

```python
plt.title("bubble chart")
plt.ylabel('salary/annum')
plt.xlabel('year')
plt.scatter(year, salary, s = population, c= color, alpha=0.9, edgecolor = "r", marker = "^")
plt.text(year[0] + 0.05, salary[0] + 0.05, population[0])
```

pie charts

```python
sizes = [3,4,6,2]
labels = ["A", "B", "C", "D"]
colors = ["blue", "red", "orange", "green"]
explode = [.2,0,0,0]
plt.title("pie chart")
plt.pie(sizes, colors = colors, explode =  explode, labels= labels, autopct= "%.5f%%", counterclock=False, startangle=90)
plt.axis("equal")
```

Histogram

```python
a = [1,2,3,4,1,3,4,2,1,1]
plt.hist(a)
xt = np.arange(23)
```

```python
b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.xticks(xt)
plt.hist(b, bins = len(b) , edgecolor = "black")     # last bin is [a,b] rest are[a,b)
```

freq of combined 

```python
plt.hist (b , bins = [1,3,6,10,15,20], edgecolor = 'red')
plt.axis([0, 25, 0,9])
```

Bargraph

```python
year       = ['year - 2012',' year - 2013',' year -2014',' year -2015',' year -2016', ' year -2017']
salary     = [12,13,14,15,16,17]
population = [100,120,180, 250,300,370]
plt.xticks(rotation = 50)
plt.bar(year, population, edgecolor = "r", alpha = 0.9)
```
